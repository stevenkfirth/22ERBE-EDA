## 4. Select only valid readings

```python
# create a list of the column names which hold the 'valid' flag
valid_columns=[f'{fuel_name}{year}Valid' for year in range(2005,2013) for fuel_name in ['Gcons','Econs']]
print(valid_columns)
```

```python
# create new dataframe (df_valid) with only rows where all Valid columns are valid (i.e =='V')
df_valid_columns=df[valid_columns]
mask=(df_valid_columns=='V').all(axis='columns')
df_valid=df[mask].copy()
df_valid
```

## 5. Create year-on-year change columns

These are added to df_valid.

```python
for year in range(2005,2012):
    
    print(year, end=' ')
    
    next_year=year+1
    
    for fuel_name in ['Gcons','Econs']:
        
        print(fuel_name, end=' ')
        
        df_valid[f'{fuel_name}{year}-{next_year}']=df_valid[f'{fuel_name}{next_year}'] - df_valid[f'{fuel_name}{year}']
        
df_valid
```

## A1. Create new dataframe with consumption data only

Doing this will make the consumption data much easier to work with.

```python
df_consumption=pd.DataFrame(columns=['HH_ID','YEAR','FUEL','VALUE','VALID_FLAG'])

for year in range(2005,2013):
    
    print(year, end=' ')
    
    for fuel_name in ['Gcons','Econs']:
        
        print(fuel_name, end=' ')
    
        df_temp=df[['HH_ID',f'{fuel_name}{year}',f'{fuel_name}{year}Valid']].copy()
        
        df_temp['YEAR']=year
        
        df_temp['FUEL']=fuel_name
        
        df_temp=df_temp.rename(
            columns={
                f'{fuel_name}{year}':'VALUE',
                f'{fuel_name}{year}Valid':'VALID_FLAG'
            }
        )
        
        df_consumption=df_consumption.append(
            df_temp,
            ignore_index=True
        )
    
df_consumption
```

## A2. Examples of analysis

```python
# selecting all valid values - where VALID_FLAG=="V"

df_consumption_valid=df_consumption[df_consumption.VALID_FLAG=="V"]
df_consumption_valid
```

```python
# calculating year-on-year changes
# 'VALUE' here is the difference between current year and next year (i.e. the next year value minus the current year value)
# takes a while to run...

df_consumption_changes=pd.DataFrame(columns=['HH_ID','YEAR_INTERVAL','FUEL','VALUE'])

for year in range(2005,2012):
    
    print(year, end=' ')
    
    next_year=year+1
    
    for fuel_name in ['Gcons','Econs']:
        
        print(fuel_name, end=' ')
        
        # current year
        df_temp1=df_consumption_valid[(df_consumption_valid.YEAR==year) & (df_consumption_valid.FUEL==fuel_name)].copy()
        df_temp1=df_temp1.set_index('HH_ID')
        
        # next year
        df_temp2=df_consumption_valid[(df_consumption_valid.YEAR==next_year) & (df_consumption_valid.FUEL==fuel_name)].copy()
        df_temp2=df_temp2.set_index('HH_ID')
        
        # join of current year and next year
        df_temp3=df_temp1.join(
            other=df_temp2,
            how='inner',
            lsuffix='1',
            rsuffix='2'
        )
        
        df_temp3=df_temp3.reset_index()
        df_temp3['YEAR_INTERVAL']=f'{year}_{next_year}'
        df_temp3['FUEL']=fuel_name
        df_temp3['VALUE']=df_temp3.VALUE2-df_temp3.VALUE1
        
        df_consumption_changes=df_consumption_changes.append(
            df_temp3[['HH_ID','YEAR_INTERVAL','FUEL','VALUE']],
            ignore_index=True
        )
        
df_consumption_changes
```
