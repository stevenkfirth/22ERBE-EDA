## run energyplus function

```python
def run_energyplus(weather_fp,input_fp,output_prefix):
    """
    Runs an energyplus simulation
    """
    !C:\Users\cvskf\EnergyPlusV9-6-0\energyplus \
        -w "$weather_fp" \
        -r \
        -p "$output_prefix" \
        "$input_fp"
        
```

## create multiple epJSON files

For each 45 deg building orientation

```python
for x in range(0,360,45):
    
    print(x)
    
    epjson['Building']['Simple One Zone (Wireframe DXF)']['north_axis']=x
    
    with open(f'1ZoneUncontrolled-{x}.epJSON','w') as f:
        json.dump(epjson,f,indent=4)
    
```

## run simulations

```python
for x in range(0,360,45):
    
    print(x)
    
    weather_fp='USA_CA_San.Francisco.Intl.AP.724940_TMY3.epw'
    
    input_fp=f'1ZoneUncontrolled-{x}.epJSON'
    
    run_energyplus(weather_fp,input_fp,x)
```
